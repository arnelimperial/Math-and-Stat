# Linear System

## [Linear Equation](https://www.encyclopediaofmath.org/index.php/Linear_equation)
> A first-order equation comprising of two variables: its graph is a straight line in the Cartesian coordinate system.


$$Ax = b$$

> where ___A___ is a linear operator, ___x___ is an unknown element and ___b___ is a given element.

## System of Linear Equation
> Group of two or more linear equations involving the same set of variables. To isolate and solve for ___x___, get the inverse of ___A___ and multiply that with ___b___.

#### _Sample Problem 1: Two simple linear equations with 2 sets variables s and t collectively as x._<br><br>
$$3s + t = 20$$<br>
$$4s - t = 8$$<br><br>
$$A = \begin{bmatrix} 3 & 1 \\ 4 & -1 \\\end{bmatrix}$$<br><br>
$$x = \begin{bmatrix} s \\ t \\\end{bmatrix}$$<br><br>
$$b = \begin{bmatrix} 20 \\ 8 \\\end{bmatrix}$$<br><br>
$$\begin{bmatrix} 3 & 1 \\ 4 & -1 \\\end{bmatrix} \begin{bmatrix} s \\ t \\\end{bmatrix} = \begin{bmatrix} 20 \\ 8 \\\end{bmatrix}$$<br><br>

> Output from [numpy](http://www.u.arizona.edu/~erdmann/mse350/topics/basic_linear_algebra.html):<br>
- s == 4 
- t == 8

$$2y + 5z = -4$$<br>
$$2x + 5y - z = 27$$<br><br>

> Output from Numpy:<br>
- x == 5
- y == 3

#### [Sample Problem 3:  Slope-intercept form](https://www.youtube.com/watch?v=SOwh-Wk7Cq0&list=PL0G-Nd0V5ZMofzthnMWaoBRq_b-5wI8PX&index=6)<br>
> Standard form example:
$$3x + 5y = 15$$<br><br>
***
> Slope-intercept form
$$5y = -3x + 15$$<br>
$$y = \frac{-3x + 15} {5}$$<br>
$$y = \frac{-3x }{5} + 3$$<br>
where slope(m) == -3/5,  $\frac {\Delta y} {\Delta x}$ (rise/run) and y-intercept == 3 or (0, 3). Look also for the x-intercept where y == 0 and points to (5, 0). 

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
# Create a dataframe with an x column containing values from -10 to 10
df = pd.DataFrame ({'x': range(-10, 11)})

# Add a y column by applying the solved equation to x
df['y'] = (-3*df['x'] / 5) + 3

#Display the dataframe
df
```

```python
%matplotlib inline

plt.plot(df.x, df.y, color="purple", marker="o")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
plt.plot(df.x, df.y, color="purple", marker="o", label='y = -3x/5 + 3')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.legend(loc='lower right')
## add axis lines for 0,0
plt.axhline()
plt.axvline()
plt.show()
```

```python
plt.plot(df.x, df.y, color="purple", marker="o")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.axhline()
plt.axvline()
plt.annotate('x-intercept', (5, 0))
plt.annotate('y-intercept', (0, 3))
plt.show()
```

```python
%matplotlib inline
#Create column for x with range(-10, 11)
df = pd.DataFrame({'x': range(-10, 11)})

#Define slope and y-intercept y = mx + b using y = -3x/5 + 3
m = -0.6
b = 3
# Add a y column by applying the slope-intercept equation to x 
df['y'] = m*(df['x']) + b
plt.plot(df.x, df.y, color="green", marker="o")
plt.xlabel('X')
plt.ylabel('Y')

plt.axhline()
plt.axvline()

plt.annotate('x-intercept', (5, 0))

#Label y-intercept
plt.annotate('y-intercept', (0, b))

# plot the slope from the y-intercept for 1x
mx = [0,1]
my = [b, b + m]
plt.plot(mx,my, color='red', lw=5)

plt.show()
```

```python
#Problem 4: Solve for system of linear equations

# x + y = 9 and 5x +15y = 75
A = np.array([[1, 1], [15, 5]])
#print(A)
B = np.array([9, 75])
#print(B)
X = np.linalg.solve(A, B)
print("Value of x, y respectively: ", X)
```

```python
#Problem 5: Solve the equation 2x -3y = 8
# slope-intercept: y = (-2x + 8) / 5
%matplotlib inline
df = pd.DataFrame({'x': range(-10, 11)})
df['y'] = (-2 *df['x']/-3)  + 8 /-3

# Find the slope(m) and y-intercept(b) of the best fit line
m, b = np.polyfit(df.x, df.y, 1)
df
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(df.x, df.y, color="blue", marker="o")
plt.grid()
plt.axhline()
plt.axvline()
plt.annotate('X-intercept', (4,0), color="red")

plt.annotate('Y-intercept', (0, b), color="black")
# plot the slope from the y-intercept for 1x
mx = [0,1]
my = [b, b + m]
plt.plot(mx,my, color='red', lw=5)


plt.show()
```

```python
def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
    
V = abline(m, b)
V
```
